```python
import pandas as pd


# Wczytujemy dane za pomocą biblioteki pandas
_100_words = pd.read_csv("_100_words_dict/accuracy_data.csv")
_500_words = pd.read_csv("_500_words_dict/accuracy_data.csv")
_10000_words = pd.read_csv("_10000_words_dict/accuracy_data.csv")


# Wyświetlamy wczytane dane
print(_100_words)
```

```python
df_100_words = pd.DataFrame(_100_words)
df_500_words = pd.DataFrame(_500_words)
df_10000_words = pd.DataFrame(_10000_words)

# Łączymy dane w jedną ramkę danych
all_values = pd.concat([df_100_words, df_500_words['accuracy'], df_10000_words['accuracy']], axis=1, keys=['_100_words', '_500_words', '_10000_words'])

# Wyświetlamy wynik
print(all_values)
```

```python
import matplotlib.pyplot as plt


# Rysujemy wykres
plt.figure(figsize=(12, 8))

# Wykres punktowy dla _100_words
plt.scatter(df_100_words['test set'], df_100_words['accuracy'], label='100')

# Wykres punktowy dla _500_words
plt.scatter(df_500_words['test set'], df_500_words['accuracy'], label='500')

# Wykres punktowy dla _10000_words
plt.scatter(df_10000_words['test set'], df_10000_words['accuracy'], label='10000')

# Dodajemy legendę, oznaczenia osi, tytuł
plt.legend(title='Legend')
plt.xlabel('Test Set')
plt.ylabel('Accuracy')
plt.title('Comparison of Different Test Sets')

# Obracamy etykiety na osi x, aby były bardziej czytelne
plt.xticks(rotation=45, ha='right')

# Wyświetlamy wykres
plt.tight_layout()
plt.show()
```

```python
import numpy as np

_100_words_npy = np.load("_100_words_dict/codebook.npy")
_500_words_npy = np.load("_500_words_dict/codebook.npy")
_10000_words_npy = np.load("_10000_words_dict/codebook.npy")

# Wypisz zawartość
print(len(_100_words_npy))
print(len(_500_words_npy))
print(len(_10000_words_npy))

def draw(data):
    x_values = data[:, 0]  # Pierwsza kolumna to współrzędna x
    y_values = data[:, 1]  # Druga kolumna to współrzędna y

    # Rysuj wykres punktowy
    plt.figure(figsize=(10, 6))
    plt.scatter(x_values, y_values, marker='o', color='b')
    plt.title(f'{len(data)} cech')
    plt.xlabel('Współrzędna X')
    plt.ylabel('Współrzędna Y')
    plt.grid(True)
    plt.show()

draw(_100_words_npy)
draw(_500_words_npy)
draw(_10000_words_npy)

```
